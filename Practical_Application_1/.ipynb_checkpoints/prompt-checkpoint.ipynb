### Required Assignment 5.1: Will the Customer Accept the Coupon?

**Context**

Imagine driving through town and a coupon is delivered to your cell phone for a restaurant near where you are driving. Would you accept that coupon and take a short detour to the restaurant? Would you accept the coupon but use it on a subsequent trip? Would you ignore the coupon entirely? What if the coupon was for a bar instead of a restaurant? What about a coffee house? Would you accept a bar coupon with a minor passenger in the car? What about if it was just you and your partner in the car? Would weather impact the rate of acceptance? What about the time of day?

Obviously, proximity to the business is a factor on whether the coupon is delivered to the driver or not, but what are the factors that determine whether a driver accepts the coupon once it is delivered to them? How would you determine whether a driver is likely to accept a coupon?

**Overview**

The goal of this project is to use what you know about visualizations and probability distributions to distinguish between customers who accepted a driving coupon versus those that did not.

**Data**

This data comes to us from the UCI Machine Learning repository and was collected via a survey on Amazon Mechanical Turk. The survey describes different driving scenarios including the destination, current time, weather, passenger, etc., and then ask the person whether he will accept the coupon if he is the driver. Answers that the user will drive there ‘right away’ or ‘later before the coupon expires’ are labeled as ‘Y = 1’ and answers ‘no, I do not want the coupon’ are labeled as ‘Y = 0’.  There are five different types of coupons -- less expensive restaurants (under \$20), coffee houses, carry out & take away, bar, and more expensive restaurants (\$20 - $50).

**Deliverables**

Your final product should be a brief report that highlights the differences between customers who did and did not accept the coupons.  To explore the data you will utilize your knowledge of plotting, statistical summaries, and visualization using Python. You will publish your findings in a public facing github repository as your first portfolio piece.





### Data Description
Keep in mind that these values mentioned below are average values.

The attributes of this data set include:
1. User attributes
    -  Gender: male, female
    -  Age: below 21, 21 to 25, 26 to 30, etc.
    -  Marital Status: single, married partner, unmarried partner, or widowed
    -  Number of children: 0, 1, or more than 1
    -  Education: high school, bachelors degree, associates degree, or graduate degree
    -  Occupation: architecture & engineering, business & financial, etc.
    -  Annual income: less than \\$12500, \\$12500 - \\$24999, \\$25000 - \\$37499, etc.
    -  Number of times that he/she goes to a bar: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    -  Number of times that he/she buys takeaway food: 0, less than 1, 1 to 3, 4 to 8 or greater
    than 8
    -  Number of times that he/she goes to a coffee house: 0, less than 1, 1 to 3, 4 to 8 or
    greater than 8
    -  Number of times that he/she eats at a restaurant with average expense less than \\$20 per
    person: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    -  Number of times that he/she goes to a bar: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    

2. Contextual attributes
    - Driving destination: home, work, or no urgent destination
    - Location of user, coupon and destination: we provide a map to show the geographical
    location of the user, destination, and the venue, and we mark the distance between each
    two places with time of driving. The user can see whether the venue is in the same
    direction as the destination.
    - Weather: sunny, rainy, or snowy
    - Temperature: 30F, 55F, or 80F
    - Time: 10AM, 2PM, or 6PM
    - Passenger: alone, partner, kid(s), or friend(s)


3. Coupon attributes
    - time before it expires: 2 hours or one day

**Investigating the Bar Coupons**

Now, we will lead you through an exploration of just the bar related coupons.  

1. Create a new `DataFrame` that contains just the bar coupons.


```python
df_bar=data_filled.query('coupon=="Bar"')
```

2. What proportion of bar coupons were accepted?


```python
ax = sns.histplot(data=df_bar,x='Y',stat="probability")

# Add labels
plt.title("Proportion of Bar Coupons Accepted")
plt.xlabel("Accepted")
plt.ylabel("Proportion")
plt.savefig("images/bar_coupons_hist_plot.png")
```

```python
bar_accpt_rate = round((len(df_bar.query(' Y == 1 ')) / len(df_bar))*100,2)
print(f'The portion of bar coupons accepted is: {bar_accpt_rate} %')
```

3. Compare the acceptance rate between those who went to a bar 3 or fewer times a month to those who went more.


```python
#first step to identify those who went to a bar 3 or fewer times
df_bar_lt3=df_bar.query('(Bar == "never") or (Bar == "less1") or (Bar == "1~3")')
# identify the acceptance in above data frame
df_bar_lt3_accpt=df_bar_lt3.query('Y == 1')
#2nd step is to identify those who went to a bar 4 or more times
df_bar_gt3=df_bar.query('(Bar == "4~8") or (Bar == "gt8")')
# identify acceptance in above data frame
df_bar_gt3_accpt=df_bar_gt3.query('Y == 1')
df_bar_accpt=len(df_bar_lt3_accpt) + len(df_bar_gt3_accpt)
#calculate acceptance rate and round it to 2 digits
bar_accpt_rate_lt3=round((len(df_bar_lt3_accpt) / df_bar_accpt)*100,2)
print(f'The acceptance rate between those who went to a bar 3 or fewer times a month to those who went more is : {bar_accpt_rate_lt3} %')
```

```python
ax1 = sns.histplot(
    data=df_bar,
    x="Bar",
    hue="Y",
    stat="probability",
    multiple="dodge",
    shrink=0.8
)
plt.title("Bar Coupon Acceptance by Frequency of Bar Visits")
plt.xlabel("Bar Visits Frequency")
plt.ylabel("Proportion Accepted")
plt.savefig("images/bar_coupon_freq_bar_visits.png")
```

4. Compare the acceptance rate between drivers who go to a bar more than once a month and are over the age of 25 to the all others.  Is there a difference?


```python
#first create a data frame where folks went to a bar at least once a month and are over 25
df_bar_month_25=df_bar.query(' (Bar == "1~3") or (Bar == "gt8") or (Bar == "4~8") ')
df_bar_month_25_accpt=df_bar_month_25.query('Y == 1 ' and 'age > 25')
df_bar_month_lt_25_accpt=df_bar_month_25.query('Y == 1 ' and 'age < 25')
df_bar_month_25_accpt_total=len(df_bar_month_25_accpt) + len(df_bar_month_lt_25_accpt)
over_25_accpt_rate = round((len(df_bar_month_25_accpt) / df_bar_month_25_accpt_total)*100,2)
under_25_accpt_rate = round((len(df_bar_month_lt_25_accpt) / df_bar_month_25_accpt_total)*100,2)
print(f'The acceptance rate of those who went to a bar more than once a month and over the age of 25 is : {over_25_accpt_rate} %')
print(f'The acceptance rate of those who went to a bar more than once a month and under the age of 25 is : {under_25_accpt_rate} %')
print("Yes, there is a very large difference")
```

5. Use the same process to compare the acceptance rate between drivers who go to bars more than once a month and had passengers that were not a kid and had occupations other than farming, fishing, or forestry.


```python
# Utilizing the data frame from above which contains folks went to a bar at least once a month
df_bar_accept_p1 = df_bar_month_25.query('(Y ==1 ) and (passanger !=  "Kid(s)")')
df_bar_did_not_accept=df_bar_month_25.query('(Y ==0 ) and (passanger !=  "Kid(s)")')
df_bar_accept_p1_total = len(df_bar_accept_p1) + len(df_bar_did_not_accept)
df_bar_accept_pm_rate = round((len(df_bar_accept_p1) / df_bar_accept_p1_total)*100,2)
print(f'The acceptance rate between drivers who go to bars more than once a month and had passengers that were not a kid and had occupations other than farming, fishing, or forestry is: {df_bar_accept_pm_rate}%')
```

### Independent Investigation

Using the bar coupon example as motivation, you are to explore one of the other coupon groups and try to determine the characteristics of passengers who accept the coupons.  

```python
# creating a data frame for restaurant less than 20 coupon type
df_restaurant_total=data_filled.query('coupon == "Restaurant(<20)"')
df_restaurant = data_filled.query('(coupon == "Restaurant(<20)") and (Y ==1)')
```

```python
#plotting the acceptance proportion
ax = sns.histplot(data=df_restaurant_total,x='Y',stat="probability")

# Add labels
plt.title("Proportion of Restaurant(<20) Accepted")
plt.xlabel("Accepted")
plt.ylabel("Proportion")
plt.savefig("images/restaurant_coupon_acceptance.png")
```

The acceptance rate is pretty high for restaurant coupons with less than $20 value as compared to Bar coupons

```python
# plotting acceptance by marital status
ax1 = sns.histplot(data=df_restaurant,x='maritalStatus',stat="count")

# Add labels
plt.title("Acceptance by Marital status")
plt.xlabel("Marital Status")
plt.ylabel("count")
plt.xticks(rotation=45, ha='right')
plt.savefig("images/restaurant_coupon_marital_status.png")
```

Acceptance is proportionately high among single and married couples

```python
# Analyzing the acceptance by Gender and Marital Status
ax2 = sns.histplot(
    data=df_restaurant,
    x="maritalStatus",
    hue="gender",
    stat="count",
    multiple="dodge",
    shrink=0.8
)
plt.title("restaurant coupon redemption by Age and Marital Status")
plt.xlabel("Marital Status")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.savefig("images/restaurant_coupon_gender_marital_status.png")
```

The proportion of males within single category is high, while among partners (unmarried & married), the proportion of females is high.

```python
# Deriving a data frame for acceptance count by age
df_restaurant_by_age=df_restaurant.groupby("age").agg("sum")[["Y"]]
```

```python
# Deriving a data frame to count acceptance by age and gender
df_restaurant_by_age_gender=df_restaurant.groupby(["age","gender"]).agg("sum")[["Y"]]
```

```python
# deriving the gender ratio by age category
df_restaurant_gender_age_proportion =  df_restaurant_by_age_gender / df_restaurant_by_age
df_restaurant_gender_age_ratio = df_restaurant_gender_age_proportion.reset_index()
```

```python
# plotting the gender ratio for each age category
fig=px.bar(df_restaurant_gender_age_ratio,x="age",y="Y",color="gender", title=" Gender Ratio by Age Category")
fig.write_image("images/restaurant_coupon_gender_ratio.png")
fig.show()
```

The acceptance rate among females is high compared to males when the targeted audience is less than 30 years of age

```python
# Analyzing the influence of destination on coupon acceptance
fig = px.bar(df_restaurant, 
       x="destination", 
       title="Coupon acceptance By Destination",
       color="destination",
       color_discrete_sequence=px.colors.qualitative.Plotly, 
       color_discrete_map={'No Urgent Place': 'red', 'Work destination': 'blue', 'Home':'green'}
      )
fig.update_layout(
    #coloraxis_showscale=False,  # Hide color scale if not needed
    plot_bgcolor='white',  # Set background color
    xaxis_title="Destination",
    yaxis_title="Count"
)
fig.write_image("images/restaurant_coupon_destination.png")
fig.show()
```

The above graph shows that the acceptance is pretty high when the audience is no hurry to reach the destination

```python
#here will use the beautiful visualization of a histogram to show variations across occupation and gender
fig=px.histogram(df_restaurant, x="occupation", title="Usage By Gender and Occupation", color="gender")
fig.write_image("images/restaurant_coupon_gender_occupation.png")
fig.show()
```

The above graph shows that the this particular coupon's usage is proportionately more among unemployed and students

```python
# plotting acceptance by Car Occupancy
fig=px.histogram(df_restaurant, x="passanger", title="Usage By Car Occupancy")
fig.show()
fig.write_image("images/restaurant_coupon_occupancy.png")
```

The above graph illustrates that this coupon is accepted more when the driver is alone or with friends
